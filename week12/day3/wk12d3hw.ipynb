```python
cars = pd.read_csv("car_dekho_details.csv")
```

```python
cars
```

```python
# split the name column by " "
cars_df = cars.name.str.split(" ", expand = True)
```

```python
# take the first two parts of the name (the rest will not be required)
cars_df_reduced = cars_df.iloc[:, [0, 1]].copy()
```

```python
#concatenate these two columns into one (make_model)
cars_df_reduced["make_model"] = cars_df_reduced[0] + " " + cars_df_reduced[1]
```

```python
# merge the columns make_model with the other columns for cars
cars = pd.concat([cars_df_reduced, cars], axis=1, ignore_index=True)
```

```python
#columns names have been lost, rename them with the original names
cars.rename(columns={0:"make_model"}, inplace = True)
cars.rename(columns={1:"TBD"}, inplace = True)
cars.rename(columns={2:"year"}, inplace = True)
cars.rename(columns={3:"selling_price"}, inplace = True)
cars.rename(columns={4:"km_driven"}, inplace = True)
cars.rename(columns={5:"fuel"}, inplace = True)
cars.rename(columns={6:"seller_type"}, inplace = True)
cars.rename(columns={7:"transmission"}, inplace = True)
cars.rename(columns={8:"owner"}, inplace = True)
```

```python
cars = cars.drop(columns = ["TBD"])
```
